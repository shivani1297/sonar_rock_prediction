```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

data collectio ad data preprocessig

```python
data = pd.read_csv("/content/Copy of sonar data.csv", header =None)
data
```

```python
data.describe()
```

```python
data.shape
```

```python
data[60].value_counts()
```

```python
data.groupby(60).mean()
```

```python
from numpy.lib.shape_base import column_stack
#seperating data and lables
X= data.drop(columns=60, axis=1)
Y= data[60]
```

```python
print(X)
```

```python
print(Y)
```

trainig ad testing data

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
print(X_train, Y_train)
```

model training = logistic regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
#accuracy o traioig data
X_train_prediction = model.predict(X_train)
trainig_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("accuracy o training dataset:", trainig_data_accuracy)
```

```python
#accuracy on test dataset
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("accuracy o training dataset:", test_data_accuracy)
```
